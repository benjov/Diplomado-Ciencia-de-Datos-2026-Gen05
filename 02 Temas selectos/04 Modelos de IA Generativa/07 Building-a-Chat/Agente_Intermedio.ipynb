```python
# ========================================
# Importar librer√≠as
# ========================================
from langchain_groq import ChatGroq
from langchain.agents import initialize_agent, Tool
from langchain.agents import AgentType
from langchain_community.tools import DuckDuckGoSearchRun
from langchain_community.tools import DuckDuckGoSearchResults
from langchain_community.utilities import DuckDuckGoSearchAPIWrapper
from langchain.utilities import SerpAPIWrapper
from langchain.tools import tool
import math
```

# 2. Clave de Groq

```python
# ========================================
# Configurar la API de Groq
# ========================================

import os, getpass
#os.environ["GROQ_API_KEY"] = getpass.getpass("üîë Ingresa tu Groq API key: ")
os.environ["GROQ_API_KEY"] = "..."
```

# 3. Importar 'motor' LLM

```python
# ========================================
# Definir el modelo LLM
# ========================================
#Models: 'openai/gpt-oss-120b', 'llama-3.3-70b-versatile'
llm = ChatGroq(
    model_name='llama-3.3-70b-versatile', # cualquier modelo open-source disponible
    temperature=0, # respuestas m√°s deterministas
    max_tokens=20000,
    streaming=True                  # stream para menor latencia
)
```

# 4. Definir herramientas

```python
# Herramienta de b√∫squeda web usando DuckDuckGo
search = DuckDuckGoSearchRun()
```

```python
# Herramienta de calculadora personalizada
@tool
def calculadora(expr: str) -> str:
    """Eval√∫a expresiones matem√°ticas simples."""
    try:
        return str(eval(expr))
    except Exception as e:
        return f"Error en el c√°lculo: {e}"
```

```python
# Lista de herramientas disponibles para el agente
tools = [
    Tool(
        name="B√∫squedaWeb",
        func=search.run,
        description="Usa esta herramienta para buscar informaci√≥n en internet"
    ),
    Tool(
        name="Calculadora",
        func=calculadora.run,
        description="Usa esta herramienta para realizar c√°lculos matem√°ticos"
    )
]
```

# 5. Crear el agente

```python
#
agent = initialize_agent(
    tools,
    llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,  # Decide en qu√© orden usar herramientas
    verbose=True
)
```

# 6. Probar el agente

```python
#
print("=== Agente de IA Avanzado ===")
print(agent.run("¬øCu√°l es la capital de Jap√≥n y cu√°nto ser√≠a 23 * 56?"))
#print(agent.run("¬øCu√°l es la noticia m√°s relevante sobre M√©xico el d√≠a de hoy?, considera s√≥lo noticias sobre deporte"))
```
